## Data  Analysis and Visualization.

```python
cancelled_perc = data['is_canceled'].value_counts()
cancelled_perc
plt.figure(figsize = (4,3))
plt.title("Reservation status count")
plt.bar(["Not cancelled", "Cancelled"], data['is_canceled'].value_counts())
```

```python
plt.figure(figsize = (4,3))
ax1 = sns.countplot(x = "hotel", hue = "is_canceled", data = data,palette = "Blues" )
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=[2,1])
plt.title("Reservation status in different hotels", size =20)
plt.xlabel("Hotel")
plt.ylabel("No of reservations")
plt.legend(["not cancelled", "is_cancelled"])
plt.show()
```

```python
resort_hotel = data[data["hotel"] == "Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize = True)
resort_hotel.head(2)
```

```python
city_hotel = data[data["hotel"] == "City Hotel"]
city_hotel["is_canceled"].value_counts(normalize = True)
city_hotel.head(2)
```

```python
#resort_hotel.head(2)
resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[["adr"]].mean()
resort_hotel.head(2)
city_hotel.head(2)
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate in City and Resort Hotels", fontsize= 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = "Resort Hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label = "City Hotel")
plt.legend(["resort hotel", "city hotel"])
plt.show()
```

```python
data["month"] = data["reservation_status_date"].dt.month
plt.figure(figsize = (16,8))
plt.title("Reservation Status Per Month", fontsize= 30)
ax1 = sns.countplot(x = "month", hue = "is_canceled", data = data)
plt.ylabel("No of reservations")
plt.legend(["not cancelled", "is_cancelled"])
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.title("ADR per month", fontsize=30)
sns.barplot(x="month", y="adr", data=data[data["is_canceled"] == 1].groupby("month")[["adr"]].sum().reset_index())
plt.show()
```
